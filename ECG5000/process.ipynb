```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("data/ECG5000_TRAIN.arff", encoding="utf-8") as f:
    header = []
    for line in f:
        if line.startswith("@attribute"):
            header.append(line.split()[1])
        elif line.startswith("@data"):
            break
    train = pd.read_csv(f, header=None)
    train.columns = header

train=train.values

train_data=train[:,0:-1]
train_label=train[:,-1]
```

```python
with open("data/ECG5000_TEST.arff", encoding="utf-8") as f:
    header = []
    for line in f:
        if line.startswith("@attribute"):
            header.append(line.split()[1])
        elif line.startswith("@data"):
            break
    test = pd.read_csv(f, header=None)
    test.columns = header

test=test.values

test_data=test[:,0:-1]
test_label=test[:,-1]
```

```python
np.save("data/train.npy",train)
np.save("data/test.npy",test)
```

```python
tmp = np.arange(5)

for i,j in enumerate(train_label):
    tmp[int(j-1)] = i

for i,j in enumerate(tmp):
    plt.grid(linestyle=":",color="b")#
    plt.plot(np.arange(140),train_data[j])
    plt.title("A ECG ({})".format(i+1))
    plt.show()
```
